### Task 1. Re-train the classifier on a different set of data.

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk import NaiveBayesClassifier, classify
import random

# Preprocess the data
def process_tokens(tokens):
    stop_words = set(nltk.corpus.stopwords.words('english'))
    return [token.lower() for token in tokens if token.isalpha() and token.lower() not in stop_words]

def get_token_dict(tokens):
    return dict([token, True] for token in tokens)

def get_tweets_for_model(cleaned_tokens_list):
    return [get_token_dict(tweet_tokens) for tweet_tokens in cleaned_tokens_list]

# # Prepare the data for the model
positive_tokens_for_model = get_tweets_for_model(positive_cleaned_tokens_list)
negative_tokens_for_model = get_tweets_for_model(negative_cleaned_tokens_list)

positive_dataset = [(tweet_dict, "Positive") for tweet_dict in positive_tokens_for_model]
negative_dataset = [(tweet_dict, "Negative") for tweet_dict in negative_tokens_for_model]

dataset = positive_dataset + negative_dataset
random.shuffle(dataset)

train_data = dataset[:int(len(dataset) * 0.7)]
test_data = dataset[int(len(dataset) * 0.7):]

# Train the classifier
classifier = NaiveBayesClassifier.train(train_data)

# Evaluate the classifier
print("Accuracy is:", classify.accuracy(classifier, test_data))
print(classifier.show_most_informative_features(10))
```

### Task 2. Try to use Logistic Regression classifier instead and compare the results with Naive Bayes.

```python
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.linear_model import LogisticRegression

# Train the Logistic Regression classifier
logistic_classifier = SklearnClassifier(LogisticRegression(max_iter=1000))
logistic_classifier.train(train_data)

# Evaluate the Logistic Regression classifier
logistic_accuracy = classify.accuracy(logistic_classifier, test_data)
print("Logistic Regression Accuracy is:", logistic_accuracy)

# Compare with Naive Bayes
naive_bayes_accuracy = classify.accuracy(classifier, test_data)
print("Naive Bayes Accuracy is:", naive_bayes_accuracy)
```
