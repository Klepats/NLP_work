### Task 1.1: Using FreqDist, find the following from textN, words that are hapaxes (unique) and those that occur less than 5 times

```python
import nltk
from nltk.corpus import gutenberg
from nltk.probability import FreqDist

# Завантаження слів з тексту "Emma" Джейн Остін
textN = gutenberg.words('austen-emma.txt')

# Обчислення частотності слів
fdist = FreqDist(textN)

# Знаходження унікальних слів (hapaxes)
hapaxes = fdist.hapaxes()

# Знаходження рідкісних слів (менше 5 разів)
rare_words = [word for word, count in fdist.items() if count < 5]

# Виведення результатів
print(f"Кількість унікальних слів (hapaxes): {len(hapaxes)}")
print(f"Слова, що зустрічаються менше 5 разів: {rare_words[:20]}")
```

### Task 1.2: draw a plot showing mapping from word length to word frequency

```python
import matplotlib.pyplot as plt
from collections import defaultdict

# Обчислення частотності слів різної довжини
word_length_freq = defaultdict(int)
for word in textN:
    word_length_freq[len(word)] += 1

# Побудова гістограми
plt.figure(figsize=(10, 5))
plt.bar(word_length_freq.keys(), word_length_freq.values())
plt.xlabel('Довжина слова')
plt.ylabel('Частота')
plt.title('Залежність частоти від довжини слова')
plt.show()
```

### Task 1.3: find 10 most frequent words occurring at the end of the sentence

```python
import nltk
from nltk.probability import FreqDist
import re

text = """Alice? Alice was beginning to get very tired of sitting by her sister on the bank. And of having nothing to do: once or twice she had peeped into the book her sister was reading. But it had no pictures or conversations in it, ‘and what is the use of a book,’ thought Alice ‘without pictures or conversation?’

So she was considering in her own mind (as well as she could, for the hot day made her feel very sleepy and stupid), whether the pleasure of making a daisy-chain would be worth the trouble of getting up and picking the daisies, when suddenly a White Rabbit with pink eyes ran close by her.

There was nothing so very remarkable in that; nor did Alice think it so very much out of the way to hear the Rabbit say to itself, ‘Oh dear! Oh dear! I shall be late!’ (when she thought it over afterwards, it occurred to her that she ought to have wondered at this, but at the time it all seemed quite natural); but when the Rabbit actually took a watch out of its waistcoat-pocket, and looked at it, and then hurried on, Alice started to her feet, for it flashed across her mind that she had never before seen a rabbit with either a waistcoat-pocket, or a watch to take out of it, and, burning with curiosity, she ran across the field after it, and fortunately was just in time to see it pop down a large rabbit-hole under the hedge.
"""  

# Розділяємо текст на речення, використовуючи токенайзер nltk.sent_tokenize
sentences = nltk.sent_tokenize(text)

# Отримуємо список останніх слів у кожному реченні: знаходимо всі слова за допомогою re.findall(r'\w+'),
# беремо останнє слово ([-1]), приводимо до нижнього регістру (.lower()), і фільтруємо лише речення з словами
last_words = [re.findall(r'\w+', sentence)[-1].lower() for sentence in sentences if re.findall(r'\w+', sentence)]

# Створюємо об’єкт FreqDist для підрахунку частоти появи кожного слова в last_words
freq_dist = FreqDist(last_words)

# Отримуємо 10 найпоширеніших слів із частотно-розподіленого списку
most_common_words = freq_dist.most_common(10)
print(most_common_words)
```

### Task 2. Create custom NLTK text of your favourite (public domain) book

#### Practice using concordance, similar, dispersion_plot, FreqDist on the resulting text.

```python
# # Відкриваємо книгу
# with open('book.txt', 'r', encoding='utf-8') as f:
#     raw = f.read()

# # Токенізація
# tokens = nltk.word_tokenize(raw)
# text = nltk.Text(tokens)

import nltk
from nltk.corpus import gutenberg

# Завантаження тексту "Alice in Wonderland" Льюїса Керрола
text = nltk.Text(gutenberg.words('carroll-alice.txt'))
```

```python
text.concordance("Alice")
```

```python
# Пошук схожих слів (схожість визначається за контекстом)
text.similar("Alice")
```

```python
# Діаграма розсіювання слів у тексті
text.dispersion_plot(['Alice', 'was', 'said', 'the', 'she'])
```

```python
# Перевіряємо, чи є text об’єктом nltk.Text, і конвертуємо його в звичайний рядок, якщо так
if isinstance(text, nltk.Text):
    text = ' '.join(text)  # Конвертуємо nltk.Text у рядок, об’єднуючи всі слова пробілами

# Очищаємо текст, видаляючи пунктуацію (включаючи апострофи ‘ і ’) та приводячи до нижнього регістру
# Використовуємо регулярний вираз, щоб залишити лише слова (літери, цифри) і пробіли
cleaned_text = re.sub(r'[^\w\s]', '', text.lower())

# Розділяємо текст на слова
words = cleaned_text.split()

# Отримуємо список англійських стоп-слів
stopwords = nltk.corpus.stopwords.words('english')

# Фільтруємо слова, виключаючи стоп-слова
filtered_words = [word for word in words if word not in stopwords]

# Створюємо FreqDist на основі відфільтрованих слів
fdist = FreqDist(filtered_words)

# Виводимо 20 найпоширеніших слів (без стоп-слів)
print("\nНайпоширеніші слова (без стоп-слів):")
print(fdist.most_common(20))  # 20 найпоширеніших слів

# Створюємо стовпчиковий графік для 20 найпоширеніших слів
words, frequencies = zip(*fdist.most_common(20))  # Розпаковуємо слова та їх частоту
plt.figure(figsize=(12, 6))  # Встановлюємо розмір графіка
plt.bar(words, frequencies)  # Створюємо стовпчиковий графік
plt.title('20 найпоширеніших слів (без стоп-слів)')  # Додаємо заголовок
plt.xlabel('Слова')  # Підпис осі X
plt.ylabel('Частота')  # Підпис осі Y
plt.xticks(rotation=45, ha='right')  # Повертаємо мітки на 45 градусів для кращої читабельності
plt.tight_layout()  # Налаштовуємо макет, щоб уникнути перекриття
plt.show()  # Показуємо графік
```
