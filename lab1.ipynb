### Task 1.1: Using FreqDist, find the following from textN, words that are hapaxes (unique) and those that occur less than 5 times

```python
import nltk
from nltk.corpus import gutenberg
from nltk.probability import FreqDist

# Ð—Ð°Ð²Ð°Ð½Ñ‚Ð°Ð¶ÐµÐ½Ð½Ñ ÑÐ»Ñ–Ð² Ð· Ñ‚ÐµÐºÑÑ‚Ñƒ "Emma" Ð”Ð¶ÐµÐ¹Ð½ ÐžÑÑ‚Ñ–Ð½
textN = gutenberg.words('austen-emma.txt')

# ÐžÐ±Ñ‡Ð¸ÑÐ»ÐµÐ½Ð½Ñ Ñ‡Ð°ÑÑ‚Ð¾Ñ‚Ð½Ð¾ÑÑ‚Ñ– ÑÐ»Ñ–Ð²
fdist = FreqDist(textN)

# Ð—Ð½Ð°Ñ…Ð¾Ð´Ð¶ÐµÐ½Ð½Ñ ÑƒÐ½Ñ–ÐºÐ°Ð»ÑŒÐ½Ð¸Ñ… ÑÐ»Ñ–Ð² (hapaxes)
hapaxes = fdist.hapaxes()

# Ð—Ð½Ð°Ñ…Ð¾Ð´Ð¶ÐµÐ½Ð½Ñ Ñ€Ñ–Ð´ÐºÑ–ÑÐ½Ð¸Ñ… ÑÐ»Ñ–Ð² (Ð¼ÐµÐ½ÑˆÐµ 5 Ñ€Ð°Ð·Ñ–Ð²)
rare_words = [word for word, count in fdist.items() if count < 5]

# Ð’Ð¸Ð²ÐµÐ´ÐµÐ½Ð½Ñ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ñ–Ð²
print(f"ÐšÑ–Ð»ÑŒÐºÑ–ÑÑ‚ÑŒ ÑƒÐ½Ñ–ÐºÐ°Ð»ÑŒÐ½Ð¸Ñ… ÑÐ»Ñ–Ð² (hapaxes): {len(hapaxes)}")
print(f"Ð¡Ð»Ð¾Ð²Ð°, Ñ‰Ð¾ Ð·ÑƒÑÑ‚Ñ€Ñ–Ñ‡Ð°ÑŽÑ‚ÑŒÑÑ Ð¼ÐµÐ½ÑˆÐµ 5 Ñ€Ð°Ð·Ñ–Ð²: {rare_words[:20]}")
```

### Task 1.2: draw a plot showing mapping from word length to word frequency

```python
import matplotlib.pyplot as plt
from collections import defaultdict

# ÐžÐ±Ñ‡Ð¸ÑÐ»ÐµÐ½Ð½Ñ Ñ‡Ð°ÑÑ‚Ð¾Ñ‚Ð½Ð¾ÑÑ‚Ñ– ÑÐ»Ñ–Ð² Ñ€Ñ–Ð·Ð½Ð¾Ñ— Ð´Ð¾Ð²Ð¶Ð¸Ð½Ð¸
word_length_freq = defaultdict(int)
for word in textN:
    word_length_freq[len(word)] += 1

# ÐŸÐ¾Ð±ÑƒÐ´Ð¾Ð²Ð° Ð³Ñ–ÑÑ‚Ð¾Ð³Ñ€Ð°Ð¼Ð¸
plt.figure(figsize=(10, 5))
plt.bar(word_length_freq.keys(), word_length_freq.values())
plt.xlabel('Ð”Ð¾Ð²Ð¶Ð¸Ð½Ð° ÑÐ»Ð¾Ð²Ð°')
plt.ylabel('Ð§Ð°ÑÑ‚Ð¾Ñ‚Ð°')
plt.title('Ð—Ð°Ð»ÐµÐ¶Ð½Ñ–ÑÑ‚ÑŒ Ñ‡Ð°ÑÑ‚Ð¾Ñ‚Ð¸ Ð²Ñ–Ð´ Ð´Ð¾Ð²Ð¶Ð¸Ð½Ð¸ ÑÐ»Ð¾Ð²Ð°')
plt.show()
```

### Task 1.3: find 10 most frequent words occurring at the end of the sentence

```python
import nltk
from nltk.probability import FreqDist
import re

text = """Alice? Alice was beginning to get very tired of sitting by her sister on the bank. And of having nothing to do: once or twice she had peeped into the book her sister was reading. But it had no pictures or conversations in it, â€˜and what is the use of a book,â€™ thought Alice â€˜without pictures or conversation?â€™

So she was considering in her own mind (as well as she could, for the hot day made her feel very sleepy and stupid), whether the pleasure of making a daisy-chain would be worth the trouble of getting up and picking the daisies, when suddenly a White Rabbit with pink eyes ran close by her.

There was nothing so very remarkable in that; nor did Alice think it so very much out of the way to hear the Rabbit say to itself, â€˜Oh dear! Oh dear! I shall be late!â€™ (when she thought it over afterwards, it occurred to her that she ought to have wondered at this, but at the time it all seemed quite natural); but when the Rabbit actually took a watch out of its waistcoat-pocket, and looked at it, and then hurried on, Alice started to her feet, for it flashed across her mind that she had never before seen a rabbit with either a waistcoat-pocket, or a watch to take out of it, and, burning with curiosity, she ran across the field after it, and fortunately was just in time to see it pop down a large rabbit-hole under the hedge.
"""  

# Ð Ð¾Ð·Ð´Ñ–Ð»ÑÑ”Ð¼Ð¾ Ñ‚ÐµÐºÑÑ‚ Ð½Ð° Ñ€ÐµÑ‡ÐµÐ½Ð½Ñ, Ð²Ð¸ÐºÐ¾Ñ€Ð¸ÑÑ‚Ð¾Ð²ÑƒÑŽÑ‡Ð¸ Ñ‚Ð¾ÐºÐµÐ½Ð°Ð¹Ð·ÐµÑ€ nltk.sent_tokenize
sentences = nltk.sent_tokenize(text)

# ÐžÑ‚Ñ€Ð¸Ð¼ÑƒÑ”Ð¼Ð¾ ÑÐ¿Ð¸ÑÐ¾Ðº Ð¾ÑÑ‚Ð°Ð½Ð½Ñ–Ñ… ÑÐ»Ñ–Ð² Ñƒ ÐºÐ¾Ð¶Ð½Ð¾Ð¼Ñƒ Ñ€ÐµÑ‡ÐµÐ½Ð½Ñ–: Ð·Ð½Ð°Ñ…Ð¾Ð´Ð¸Ð¼Ð¾ Ð²ÑÑ– ÑÐ»Ð¾Ð²Ð° Ð·Ð° Ð´Ð¾Ð¿Ð¾Ð¼Ð¾Ð³Ð¾ÑŽ re.findall(r'\w+'),
# Ð±ÐµÑ€ÐµÐ¼Ð¾ Ð¾ÑÑ‚Ð°Ð½Ð½Ñ” ÑÐ»Ð¾Ð²Ð¾ ([-1]), Ð¿Ñ€Ð¸Ð²Ð¾Ð´Ð¸Ð¼Ð¾ Ð´Ð¾ Ð½Ð¸Ð¶Ð½ÑŒÐ¾Ð³Ð¾ Ñ€ÐµÐ³Ñ–ÑÑ‚Ñ€Ñƒ (.lower()), Ñ– Ñ„Ñ–Ð»ÑŒÑ‚Ñ€ÑƒÑ”Ð¼Ð¾ Ð»Ð¸ÑˆÐµ Ñ€ÐµÑ‡ÐµÐ½Ð½Ñ Ð· ÑÐ»Ð¾Ð²Ð°Ð¼Ð¸
last_words = [re.findall(r'\w+', sentence)[-1].lower() for sentence in sentences if re.findall(r'\w+', sentence)]

# Ð¡Ñ‚Ð²Ð¾Ñ€ÑŽÑ”Ð¼Ð¾ Ð¾Ð±â€™Ñ”ÐºÑ‚ FreqDist Ð´Ð»Ñ Ð¿Ñ–Ð´Ñ€Ð°Ñ…ÑƒÐ½ÐºÑƒ Ñ‡Ð°ÑÑ‚Ð¾Ñ‚Ð¸ Ð¿Ð¾ÑÐ²Ð¸ ÐºÐ¾Ð¶Ð½Ð¾Ð³Ð¾ ÑÐ»Ð¾Ð²Ð° Ð² last_words
freq_dist = FreqDist(last_words)

# ÐžÑ‚Ñ€Ð¸Ð¼ÑƒÑ”Ð¼Ð¾ 10 Ð½Ð°Ð¹Ð¿Ð¾ÑˆÐ¸Ñ€ÐµÐ½Ñ–ÑˆÐ¸Ñ… ÑÐ»Ñ–Ð² Ñ–Ð· Ñ‡Ð°ÑÑ‚Ð¾Ñ‚Ð½Ð¾-Ñ€Ð¾Ð·Ð¿Ð¾Ð´Ñ–Ð»ÐµÐ½Ð¾Ð³Ð¾ ÑÐ¿Ð¸ÑÐºÑƒ
most_common_words = freq_dist.most_common(10)
print(most_common_words)
```

```python
# ÐŸÐµÑ€ÐµÐ²Ñ–Ñ€ÑÑ”Ð¼Ð¾, Ñ‡Ð¸ Ñ” text Ð¾Ð±â€™Ñ”ÐºÑ‚Ð¾Ð¼ nltk.Text, Ñ– ÐºÐ¾Ð½Ð²ÐµÑ€Ñ‚ÑƒÑ”Ð¼Ð¾ Ð¹Ð¾Ð³Ð¾ Ð² Ð·Ð²Ð¸Ñ‡Ð°Ð¹Ð½Ð¸Ð¹ Ñ€ÑÐ´Ð¾Ðº, ÑÐºÑ‰Ð¾ Ñ‚Ð°Ðº
if isinstance(text, nltk.Text):
    text = ' '.join(text)  # ÐšÐ¾Ð½Ð²ÐµÑ€Ñ‚ÑƒÑ”Ð¼Ð¾ nltk.Text Ñƒ Ñ€ÑÐ´Ð¾Ðº, Ð¾Ð±â€™Ñ”Ð´Ð½ÑƒÑŽÑ‡Ð¸ Ð²ÑÑ– ÑÐ»Ð¾Ð²Ð° Ð¿Ñ€Ð¾Ð±Ñ–Ð»Ð°Ð¼Ð¸

# ÐžÑ‡Ð¸Ñ‰Ð°Ñ”Ð¼Ð¾ Ñ‚ÐµÐºÑÑ‚, Ð²Ð¸Ð´Ð°Ð»ÑÑŽÑ‡Ð¸ Ð¿ÑƒÐ½ÐºÑ‚ÑƒÐ°Ñ†Ñ–ÑŽ (Ð²ÐºÐ»ÑŽÑ‡Ð°ÑŽÑ‡Ð¸ Ð°Ð¿Ð¾ÑÑ‚Ñ€Ð¾Ñ„Ð¸ â€˜ Ñ– â€™) Ñ‚Ð° Ð¿Ñ€Ð¸Ð²Ð¾Ð´ÑÑ‡Ð¸ Ð´Ð¾ Ð½Ð¸Ð¶Ð½ÑŒÐ¾Ð³Ð¾ Ñ€ÐµÐ³Ñ–ÑÑ‚Ñ€Ñƒ
# Ð’Ð¸ÐºÐ¾Ñ€Ð¸ÑÑ‚Ð¾Ð²ÑƒÑ”Ð¼Ð¾ Ñ€ÐµÐ³ÑƒÐ»ÑÑ€Ð½Ð¸Ð¹ Ð²Ð¸Ñ€Ð°Ð·, Ñ‰Ð¾Ð± Ð·Ð°Ð»Ð¸ÑˆÐ¸Ñ‚Ð¸ Ð»Ð¸ÑˆÐµ ÑÐ»Ð¾Ð²Ð° (Ð»Ñ–Ñ‚ÐµÑ€Ð¸, Ñ†Ð¸Ñ„Ñ€Ð¸) Ñ– Ð¿Ñ€Ð¾Ð±Ñ–Ð»Ð¸
cleaned_text = re.sub(r'[^\w\s]', '', text.lower())

# Ð Ð¾Ð·Ð´Ñ–Ð»ÑÑ”Ð¼Ð¾ Ñ‚ÐµÐºÑÑ‚ Ð½Ð° ÑÐ»Ð¾Ð²Ð°
words = cleaned_text.split()

# ÐžÑ‚Ñ€Ð¸Ð¼ÑƒÑ”Ð¼Ð¾ ÑÐ¿Ð¸ÑÐ¾Ðº Ð°Ð½Ð³Ð»Ñ–Ð¹ÑÑŒÐºÐ¸Ñ… ÑÑ‚Ð¾Ð¿-ÑÐ»Ñ–Ð²
stopwords = nltk.corpus.stopwords.words('english')

# Ð¤Ñ–Ð»ÑŒÑ‚Ñ€ÑƒÑ”Ð¼Ð¾ ÑÐ»Ð¾Ð²Ð°, Ð²Ð¸ÐºÐ»ÑŽÑ‡Ð°ÑŽÑ‡Ð¸ ÑÑ‚Ð¾Ð¿-ÑÐ»Ð¾Ð²Ð°
filtered_words = [word for word in words if word not in stopwords]

# Ð¡Ñ‚Ð²Ð¾Ñ€ÑŽÑ”Ð¼Ð¾ FreqDist Ð½Ð° Ð¾ÑÐ½Ð¾Ð²Ñ– Ð²Ñ–Ð´Ñ„Ñ–Ð»ÑŒÑ‚Ñ€Ð¾Ð²Ð°Ð½Ð¸Ñ… ÑÐ»Ñ–Ð²
fdist = FreqDist(filtered_words)

# Ð’Ð¸Ð²Ð¾Ð´Ð¸Ð¼Ð¾ 20 Ð½Ð°Ð¹Ð¿Ð¾ÑˆÐ¸Ñ€ÐµÐ½Ñ–ÑˆÐ¸Ñ… ÑÐ»Ñ–Ð² (Ð±ÐµÐ· ÑÑ‚Ð¾Ð¿-ÑÐ»Ñ–Ð²)
print("\nÐÐ°Ð¹Ð¿Ð¾ÑˆÐ¸Ñ€ÐµÐ½Ñ–ÑˆÑ– ÑÐ»Ð¾Ð²Ð° (Ð±ÐµÐ· ÑÑ‚Ð¾Ð¿-ÑÐ»Ñ–Ð²):")
print(fdist.most_common(20))  # 20 Ð½Ð°Ð¹Ð¿Ð¾ÑˆÐ¸Ñ€ÐµÐ½Ñ–ÑˆÐ¸Ñ… ÑÐ»Ñ–Ð²

# Ð¡Ñ‚Ð²Ð¾Ñ€ÑŽÑ”Ð¼Ð¾ ÑÑ‚Ð¾Ð²Ð¿Ñ‡Ð¸ÐºÐ¾Ð²Ð¸Ð¹ Ð³Ñ€Ð°Ñ„Ñ–Ðº Ð´Ð»Ñ 20 Ð½Ð°Ð¹Ð¿Ð¾ÑˆÐ¸Ñ€ÐµÐ½Ñ–ÑˆÐ¸Ñ… ÑÐ»Ñ–Ð²
words, frequencies = zip(*fdist.most_common(20))  # Ð Ð¾Ð·Ð¿Ð°ÐºÐ¾Ð²ÑƒÑ”Ð¼Ð¾ ÑÐ»Ð¾Ð²Ð° Ñ‚Ð° Ñ—Ñ… Ñ‡Ð°ÑÑ‚Ð¾Ñ‚Ñƒ
plt.figure(figsize=(12, 6))  # Ð’ÑÑ‚Ð°Ð½Ð¾Ð²Ð»ÑŽÑ”Ð¼Ð¾ Ñ€Ð¾Ð·Ð¼Ñ–Ñ€ Ð³Ñ€Ð°Ñ„Ñ–ÐºÐ°
plt.bar(words, frequencies)  # Ð¡Ñ‚Ð²Ð¾Ñ€ÑŽÑ”Ð¼Ð¾ ÑÑ‚Ð¾Ð²Ð¿Ñ‡Ð¸ÐºÐ¾Ð²Ð¸Ð¹ Ð³Ñ€Ð°Ñ„Ñ–Ðº
plt.title('20 Ð½Ð°Ð¹Ð¿Ð¾ÑˆÐ¸Ñ€ÐµÐ½Ñ–ÑˆÐ¸Ñ… ÑÐ»Ñ–Ð² (Ð±ÐµÐ· ÑÑ‚Ð¾Ð¿-ÑÐ»Ñ–Ð²)')  # Ð”Ð¾Ð´Ð°Ñ”Ð¼Ð¾ Ð·Ð°Ð³Ð¾Ð»Ð¾Ð²Ð¾Ðº
plt.xlabel('Ð¡Ð»Ð¾Ð²Ð°')  # ÐŸÑ–Ð´Ð¿Ð¸Ñ Ð¾ÑÑ– X
plt.ylabel('Ð§Ð°ÑÑ‚Ð¾Ñ‚Ð°')  # ÐŸÑ–Ð´Ð¿Ð¸Ñ Ð¾ÑÑ– Y
plt.xticks(rotation=45, ha='right')  # ÐŸÐ¾Ð²ÐµÑ€Ñ‚Ð°Ñ”Ð¼Ð¾ Ð¼Ñ–Ñ‚ÐºÐ¸ Ð½Ð° 45 Ð³Ñ€Ð°Ð´ÑƒÑÑ–Ð² Ð´Ð»Ñ ÐºÑ€Ð°Ñ‰Ð¾Ñ— Ñ‡Ð¸Ñ‚Ð°Ð±ÐµÐ»ÑŒÐ½Ð¾ÑÑ‚Ñ–
plt.tight_layout()  # ÐÐ°Ð»Ð°ÑˆÑ‚Ð¾Ð²ÑƒÑ”Ð¼Ð¾ Ð¼Ð°ÐºÐµÑ‚, Ñ‰Ð¾Ð± ÑƒÐ½Ð¸ÐºÐ½ÑƒÑ‚Ð¸ Ð¿ÐµÑ€ÐµÐºÑ€Ð¸Ñ‚Ñ‚Ñ
plt.show()  # ÐŸÐ¾ÐºÐ°Ð·ÑƒÑ”Ð¼Ð¾ Ð³Ñ€Ð°Ñ„Ñ–Ðº
```
